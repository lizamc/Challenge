```python
import string
from datetime import datetime
from collections import Counter, defaultdict
import codecs
import csv
from urllib.parse import urlparse, parse_qs, urlsplit
import sys
```

```python
filename = 'Clickstream.csv'

with open(filename, encoding='UTF8') as f:
    clicks = f.readlines()
#no..
for x in range(len(clicks)):
    clicks[x] = clicks[x].split(",")
```

```python
#6-PageURL
#7-Referral

clicks_search = []

for click in clicks:
    if 'search/items.aspx?q=' in click[7] and click[8]!='-2':
        clicks_search.append(click)
```

```python
d = defaultdict(list)
```

```python
clicks_query = []
```

```python
print(clicks[0])
print(clicks_search[0])
```

```python
for click in clicks_search:
    URL = click[7]
    parsed_url = urlsplit(URL)
    par = parse_qs(parsed_url.query)
    print(par["q"])
    query_string=par['q'][0].strip()
    clicks_query.append(query_string)
    d[query_string].append(click[8])
    
```
